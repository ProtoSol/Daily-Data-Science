# Numpy

## Creation

```python
# creation of numpy arrays

# first we have to import the numpy library, if not present install it using pip install numpy
import numpy as np

# using this shorthand we can create numpy arrays
```

```python
# simple creation of 1D (vector) array
a = np.array([1, 2, 3, 4, 5])
print(type(a))
a
```

```python
# we can also create 2D (matrix) or 3D (tensor) arrays using nested lists
b = np.array([[1, 2, 3], [4, 5, 6]])
print(type(b))
print(b)

c = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(type(c))
print(c)
```

```python
# We can also use list or tuple comprehensions to create arrays
d = np.array([i for i in range(1, 11)])
print(d)

e = np.array([(i, i**2) for i in range(1, 6)])
print(e)
```

```python
# now let's try to pass a python dict set to the array function
my_dict = {'a': 1, 'b': 2, 'c': 3}
f = np.array(my_dict)
print(f)
type(f)

my_set = {1, 2, 2, 3, 4, 5, 5}
g = np.array(my_set)
print(g)
type(g)
```

## Functions

```python
# we can also change the data type of the array elements using dtype parameter
d = np.array([1, 2, 3, 4, 5], dtype=float)
print(d)
print(d.dtype)

e = np.array([1, 0, 3, -1, 5], dtype=bool)
print(e)
print(e.dtype)
```

```python
# arrange function to create arrays with a range of values
f = np.arange(0, 10)  # start, stop
print(f)


# if you want to use the step parameter with floating point numbers, you can use the linspace function instead
g = np.linspace(0, 1, 5)  # start, stop, steps
print(g)
```

```python
# reshape function to change the shape of an array
h = np.arange(1, 13)
print(h)
h_reshaped = h.reshape(3,4) # reshape to 3 rows and 4 columns
print(h_reshaped)
```

```python
# these functions can be used simultaneously too
np.arange(1, 13).reshape(3, 4)[0][0].dtype
```

```python
print(np.ones((2, 3))) # create a 2x3 array of ones
print()
print(np.zeros((4, 5))) # create a 4x5 array of zeros
print()
print(np.eye(3)) # create a 3x3 identity matrix
```

```python
# Now we will use the random module of numpy to create random arrays
print(np.random.rand(3, 4)) # create a 3x4 array of random numbers between 0 and 1
print()
print(np.random.randint(0, 10, (2, 5))) # create a 2x5 array of random integers between 0 and 10
```

```python
# To control the randomness we can set the seed value
np.random.seed(42)
print(np.random.rand(3, 4)) # create a 3x4 array of random numbers between 0 and 1
print()
print(np.random.randint(0, 10, (2, 5))) # create a 2x5 array of random integers between 0 and 10
```

```python
# linespace and arange functions can also be used to create arrays with specific ranges and steps
print(np.arange(0, 10, 2)) # create an array of even numbers
print()
print(np.linspace(0, 1, 5)) # create an array of 5

# linespace linearlly generates values between the start and stop values, while arange generates values with a specific step size. So if you want to create an array of 5 values between 0 and 1, you can use linspace, but if you want to create an array of even numbers between 0 and 10, you can use arange with a step size of 2.
# while arrange is more flexible and can be used to create arrays with specific step sizes, linspace is more precise and can be used to create arrays with a specific number of values between a range.
```

```python
# Simple way to create identity matrix
print(np.eye(4)) # create a 4x4 identity matrix
print(np.identity(4)) # create a 4x4 identity matrix
```

## Attributes

```python
# To find the dimensions of an array we can use the ndim attribute
print(a.ndim) # number of dimensions

# Now to find the shape of an array we can use the shape attribute
print(a.shape) # shape of the array ->> (2, 3) (where 2 means 2 rows and 3 means 3 columns)


b = np.array([[[1,2], [3, 4]], [[5, 6], [7, 8]]], dtype=np.int16)
print(b.ndim) # number of dimensions now it is 3 because it has 3 levels of nesting (blocks, rows, columns)
print(b.shape) # shape of the array ->> (2, 2, 2) (where 2 means 2 blocks, 2 means 2 rows and 2 means 2 columns)
```

```python
# Now to find the size of an array we can use the size attribute
print(a.size) # size of the array ->> 6 (total number of elements in
print(b.size) # size of the array ->> 8 (total number of elements in the array)

# This tells me the total number of elements in the array, now to find the size of the element we can use the item-size attribute
print(a.itemsize)
print(b.itemsize)
```

```python
# By the way we use dtype to find the datatype of the element
print(a.dtype)
print(b.dtype)
```

```python
# To Transpose an array we can use the T attribute
print(a.T) # transpose of the array a
print()
print(b.T) # transpose of the array b
```

`and similarly to the scalar operations we can perform all the arithmetic and relational operations using another vector.`

```python
# Let's test with different shapes
a3 = np.arange(6).reshape(2, 3)
print(a3)
print()
a4 = np.arange(6, 12).reshape(3, 2)
print(a4)
print()
try :
    print(a3 + a4) # this will give an error because the shapes are not compatible for element-wise addition
    print()
except ValueError as e:
    print(e)
    print()
print(a3.dot(a4)) # but here we can perform matrix multiplication using the dot method because the inner dimensions are compatible (3 in a3 and 3 in a4)
```

```python
# cos sin tan functions
print(np.cos(a1)) # cosine of all elements in the array a1
print(np.sin(a1)) # sine of all elements in the array a1
print(np.tan(a1)) # tangent of all elements in the array a1
```

## Index Slicing

```python
a1 = np.arange(10)
a2 = np.arange(12).reshape(3, 4)
a3 = np.arange(8).reshape(2, 2, 2)
```

```python
print(a1)
print()
print(a2)
print()
print(a3)
```

`To perform the indexing and slicing operations on the arrays, we can use the standard Python indexing and slicing syntax.
`

```python
# Indexing 

print(a1[0]) # first element of the array a1
print(a2[1][0]) # first element of the first row of the array a2

# We can use another way to access the elements of the array using a single set of square brackets and comma-separated indices
print(a2[0, 3]) # first element of the first row of the array
print(a3[0, 1, 0]) # first element of the first block, first row and first column of the array a3
```

```python
# slicing operations

print(a1[0:2]) # first two elements of the array a1
print()
print(a2[0:2, 1:3]) # first two rows and columns 1 and 2 of the array a2
print()
print(a3[0:1, 0:2, 0:2]) # first block, first two rows and first two columns of the array a3
print()

# We can also use negative indexing or steps in slicing
print(a1[-2:]) # last two elements of the array a1
print()
print(a2[::2, ::2]) # every other row and every other column of
```

## Stacking

```python
# We can stack arrays vertically or horizontally using the vstack and hstack functions
a2 = np.arange(12).reshape(3, 4)
print(a2)
print()

a3 = np.arange(12).reshape(3, 4)
print(a3)
print()

a4 = np.vstack((a2, a3)) # stack a2 and a3 vertically
print(a4)
print()

a5 = np.hstack((a2, a3)) # stack a2 and a3 horizontally
print(a5)
```

```python
# Now what if we try to stack arrays of different shapes, we will get an error because the shapes are not compatible for stacking
a6 = np.arange(6).reshape(2, 3)
print(a6)
print()
try:
    a7 = np.vstack((a2, a6)) # this will give an error because the shapes are not compatible for vertical stacking
    print(a7)
except ValueError as e:
    print(e)
    print()
```

## Splitting

```python
# Now we will use the splitting functions to split arrays into smaller arrays
a8 = np.arange(12).reshape(3, 4)
print(a8)
print()
a9 = np.hsplit(a8, 2) # split the array a8 into
print(a9) # this will split the array a8 into 2 equal parts horizontally
print()
a10 = np.vsplit(a8, 3) # split the array a8 into 3 equal parts vertically
print(a10) # this will split the array a8 into 3 equal parts vertically

a10[1].dtype # thus a10 is a list of 3 arrays, each of shape (1, 4) and dtype int64
```
