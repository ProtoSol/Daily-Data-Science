## Null Values in the Dataset

```python
df.isnull().sum()
```

Percentage of Null values in the Dataset -

```python
df.isnull().sum() * 100 / len(df)
```

If any column has more than 70 to 75% data as NULL or NA, then it is not useful in extracting insights.\
But it depends on the Scenario.\
If the values is neglible like for the Embarked_town and Embark column we remove the rows, as It dosnt effect much on the result.\
For other Columns we can impute the values by various methods.\
For eg. Replacing with mean or median, if catagorical data then we replace with the mode.

## Replacing Null values with Mean values

```python
print(df['age'].min())
print(df['age'].mean())
print(df['age'].max())
```

```python
df['age'].unique()
```

```python
df2 = df['age'].fillna(df['age'].mean())
df2
```

```python
df2.isnull().sum()
```

```python
df2.unique()
```

## Replacing Null values with Mode values

```python
df.isnull().sum()
```

```python
df3 = df # Copying the dataset
```

```python
df3 = df3.drop(columns = ['deck'], axis = 1)
```

```python
df3.isnull().sum()
```

```python
df3['embark_town'].fillna(df3['embark_town'].mode()[0], inplace = True)
```

```python
df3['embarked'].fillna(df3['embarked'].mode()[0], inplace = True)
```

```python
df3.isnull().sum()
```

# Applying to the Original Dataset

```python
df_updated = df
df_updated
```

```python
df_updated['age'].fillna(df_updated['age'].mean(), inplace = True)
```

```python
df_updated.drop(columns = ['deck'], axis = 1, inplace = True)
```

```python
df_updated['embarked'].fillna(df_updated['embarked'].mode()[0], inplace = True)
```
