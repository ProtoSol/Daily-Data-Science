Representation of Rows and Column. The Row and Column can have different representation terminologies, such as -

Row -> Records, Observations, Instances, Entries, Datapoints.

Columns -> Attributes, Dimensions, Variables, Properties, Features, Fields.

Rows and Columns are Important as they are a part of Data Frame that is a type of Data Structure of Pandas.
They are a type of Data set representation just like CSV (comma seperated data), TSV (tab seperated data).

Levels of Measurement :->

- Nominal
- Ordinal
- Interval
- Ratio

Nominal (Only Names) -> Ordinal (Ordered Name) -> Interval (Name + Order + Proportion / Intervals) -> Ratio (Interval and Contains Absolute Zero)
