## Types of Plots ->

1. Scatter Plot

  It is used to plot relationship between two variables. Both values are numerical not categorical.

2. Bar Plot / Chart

  It is used to plot single numerical Variable using the mean and shows the standard deviation scale.

3. Box Plot

  It is used to observe the distribution of data, and possible outliers.

4. Pie Chart / Donut Chart

  As the name suggests it shows the percentage of categorical values.

5. Area Chart

  It shows the data in the form of Area coverage.

6. Spider Plot

  Like a cobweb, it starts from center and spreads out according to the value.

7. Geographical Plot

  It uses Map to plot value on the certain area,

8. Guide to Choose Types of Plot -> [Andrew Abela Chart Guide](https://extremepresentation.typepad.com/files/choosing-a-good-chart-09.pdf)

## Code for Plotting Data ->

```python
# Import the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the Dataset
df = sns.load_dataset('titanic')
```

```python
# Ploting a Scatter Plot
df.plot(kind='scatter', x='age', y='fare', figsize=(10, 5), color = 'red')

# The color argument can accept color name and hexcode of the color for eg #00ecfc.
```

```python
# Code for Bar Chart

# We can also plot using the groupby function
df.groupby('sex')['fare'].mean().plot(kind='bar', color='#00ecfc')
plt.ylabel('Average Fare')
plt.xlabel('Gender')
plt.title('Average Fare by Gender')
```

```python
# Plots by seaborn library

# Bar plot using seaborn
sns.barplot(x='survived', y='fare', data=df, hue='class')
```

```python
# Box plot using seaborn
sns.boxplot(x='sex', y='age', data=df, hue='class')
```

```python
# Scatter plot using Seaborn
sns.scatterplot(x='fare', y='age', data=df, hue='class')
plt.title('Age v/s Fare')
```

```python
# Interactive plot using ploty
import plotly.express as px
fig = px.scatter(df, x='age', y='fare', color='class')
fig.show()
```

```python
# Interactive Box plot
fig = px.box(df, x='sex', y='fare', color='survived')
fig.show()
```

```python
# Example of advance plot of Plotly
import plotly.graph_objects as go
import numpy as np
np.random.seed(1)

N = 70

fig = go.Figure(data=[go.Mesh3d(x=(70*np.random.randn(N)),
                   y=(55*np.random.randn(N)),
                   z=(40*np.random.randn(N)),
                   opacity=0.5,
                   color='rgba(244,22,100,0.6)'
                  )])

fig.update_layout(
    scene = dict(
        xaxis = dict(nticks=4, range=[-100,100],),
                     yaxis = dict(nticks=4, range=[-50,100],),
                     zaxis = dict(nticks=4, range=[-100,100],),),
    width=700,
    margin=dict(r=20, l=10, b=10, t=10))

fig.show()
```
