>**To Convert Size, Installs and Price into Numeric Form**

```python
## Check for Na values
df['Size'].isna().sum()
```

```python
df['Installs'].isna().sum()
```

```python
df['Price'].isna().sum()
```

```python
df.head()
```

```python
# Check Unique Values of the Size Column
df['Size'].unique()
```

```python
# Convert the {number}K into {converted number}M
df['Size'] = df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN'))
df['Size'] = df['Size'].apply(lambda x: str(x).replace('M', ''))
df['Size'] = df['Size'].apply(lambda x: float(x.replace('k', '')) / 1000 if 'k' in x else float(x))
df.head()
```

```python
# Now we move to Installs
df['Installs'].value_counts()
```

```python
# Group the values of Installs column using binning process
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+', ''))
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace(',', ''))
```

```python
df.head()
```

```python
# Now we move to the Prices where we Remove the Dollar Sign
df['Price'].unique()
```

```python
df['Price'] = df['Price'].apply(lambda x: str(x).replace('$', ''))
df.head()
```

```python
df['Price'].unique()
```

```python
df['Installs'] = df['Installs'].astype(int)
df['Price'] = df['Price'].astype(float)
```
