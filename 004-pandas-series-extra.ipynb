# Series (More Functions)

```python
# astype function for Pandas

import sys

print(kohli.info()) # This is using 3.5 KB of memory
print()
print("Now changing the datatype of the dataset col. :")
kohli = kohli.astype({'match_no': 'int8', 'runs': 'int16'}) # This now uses 777 B of memory
print(kohli.info())
```

```python
# between function for Pandas
print(kohli[kohli['runs'].between(50, 100)].size) # This will return a boolean series with True for values between 50 and 100 and False for others 
# So there are 94 values between 50 and 100 in the runs column of the kohli dataset
```

```python
# clip function for Pandas
print(kohli['runs'].clip(50, 100)) # This will replace all values less than 50 with 50 and all values greater than 100 with 100 in the
# runs column of the kohli dataset
```

```python
# drop duplicates function for Pandas
print(kohli['runs'].drop_duplicates()) # This will return a series with only unique values in the runs column of the kohli dataset
# there is also keep parameter in the drop_duplicates function which can be used to keep the first, last or all duplicates in the series.

print()

# We can use .duplicated function to find out which values are duplicated in the series. It will return a boolean series with True for duplicated values and False for unique values.
# print(kohli['runs'].duplicated()) # This will return a boolean series with True for duplicated values and False for unique values in the runs column of the kohli dataset
print(kohli['runs'].duplicated().sum()) # This will return the number of duplicated values in the runs column of the kohli dataset
```

```python
# isnull function for Pandas
kohli['runs'].isnull() # This will return a boolean series with True for null values and False for non-null values in the runs column of the kohli dataset
print(kohli['runs'].isnull().sum()) # This will return the number of null values in the runs column of the kohli dataset
```

```python
# dropna function for Pandas
kohli['runs'].dropna() # This will return a series with only non-null values in the runs column of the kohli dataset
print(kohli['runs'].dropna().size) # This will return the number of non-null values in the runs column of the kohli dataset
```

```python
# fillna function for Pandas
kohli['runs'].fillna(0) # This will replace all null values with 0 in the runs column of the kohli dataset
print(kohli['runs'].fillna(0).size) # This will return the number of values in the runs column of the kohli dataset after replacing null values with 0

# forward fill function for Pandas
kohli['runs'].fillna(method='ffill') # This will replace all null values with the previous non-null value in the runs column of the kohli dataset
print(kohli['runs'].fillna(method='ffill').size) # This will return the number of values in the runs column of the kohli dataset after replacing null values with the previous non-null value

# as we can see we will now use ffill or bfill method to fill the null values in the runs column of the kohli dataset instead of filling it with 0 as it will be more appropriate to fill it with the previous or next non-null value rather than filling it with 0 which will be misleading in the analysis.
```

```python
# isin function for Pandas
print(kohli['runs'].isin([0, 50, 100])) # This will return a boolean series with True for values that are in the list [0, 50, 100]
```

```python
# apply function for Pandas
print(kohli['runs'].apply(lambda x: x*2)) # This will return a series with all values in the runs column of the kohli dataset multiplied by 2

# this can be used for one hot encoding of categorical variables in the dataset. For example if we have a column with values 'A', 'B' and 'C' we can use the apply function to create three new columns with values 1 and 0 for each category. This is called one hot encoding and it is used to convert categorical variables into numerical variables which can be used for machine learning algorithms.
```

```python
# copy function for Pandas

# this is used to create deepcopy of the series or dataframe. It is used to create a new object with the same data as the original object but with a different memory location. This is useful when we want to make changes to the new object without affecting the original object.

kohli_head = kohli.head() # This will return the first 5 rows of the kohli dataset but if we apply any changes to kohli_head it will be applied to the kohli dataset. kohli_head here is a view of the kohli dataset and not a copy of it. So if we want to make changes to the kohli_head without affecting the kohli dataset we can use the copy function to create a deep copy of the kohli_head.
# as this is a shallow copy of the kohli dataset. So if we want to make changes to the kohli_head without affecting the kohli dataset we can use the copy function to create a deep copy of the kohli_head.

kohli_head_copy = kohli_head.copy() # This will create a deep copy of the kohli_head dataset. So if we apply any changes to kohli_head_copy it will not affect the kohli dataset.
```
