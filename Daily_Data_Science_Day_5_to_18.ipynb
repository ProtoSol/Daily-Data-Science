## Filtering a Dataset

```python
df.head()
```

```python
# To see the unique values in a Column we will use .unique function.
df.sex.unique()
```

```python
# To filter values of the dataset we can use the indexing
df[(df.sex == 'female')]
```

```python
# Similiarly
df.embark_town.unique()
```

```python
# We can use this to get the data acc to the conditions.
df[(df.embark_town == 'Southampton')]
```

```python
# To get the number of the females from Southhampton we can use the following code.
df[(df.embark_town == 'Southampton') & (df.sex == 'female')]
# Each condition should be in a seperate bracket
```

```python
# We can extend the query
df[((df.embark_town == 'Southampton') | (df.embark_town == 'Queenstown')) & (df.sex == 'female')]
```

```python
# Other way of writing the function is using the inbuilt function of the dataframe.
df[df.embark_town.isin(['Queenstown', 'Southampton']) & (df.sex == 'female')].head()
```

> Thus there are many ways to do the same task in the python. Also on the side note, we perform complex condtional operations.

## Splitting a String into multiple columns

```python
import pandas as pd

df = pd.DataFrame({'name':['Pranshu Saini', 'Mayank Shingari', 'Gaurav Chaubey', 'Prabhat Kumar'],
                   'location':['Patiala, Punjab', 'Dehradun, Uttrakhand', 'Faridabad, Haryana', 'Patna, Bihar']})
df
```

```python
# Now to split the string of one column to multiple columns we can use.

df.name.str.split(' ', expand = True)
```

```python
df[["first_name", "second_name"]] = df.name.str.split(' ', expand = True)
# Now we will store the values in the original dataframe

df
```

```python
# For the location we can do the same
df[["city", "state"]] = df.location.str.split(', ', expand = True)
df
```

```python
# And updating the dataframe by droping the old columns
df = df.drop(['name', 'location'], axis = 1)
df
```

## Reshaping the Dataframe

```python
locations = pd.DataFrame([['12345',100,200,300], ['34567', 400, 500, 600], ['67890', 700, 800, 900]],
                         columns = ['zip', 'factory', 'warehouse', 'retail'])
locations.head()
```

```python
locations.head().T
```

> To make it more readable and easy to analyze we use .melt() function

```python
locations_long = locations.melt(id_vars = 'zip', var_name = 'location_type', value_name = 'distance')
# Syntax of the melt function
locations_long
```

```python
sns.barplot(x = 'zip', y = 'factory', data = locations)
```

```python
sns.barplot(x = 'zip', y = 'distance', hue = 'location_type', data = locations_long)
```

```python
sns.barplot(x = 'zip', y = 'factory', data= locations)
```

Thus we see that the melt function was used to melt other parameters and plot a single graph, that has the information for all the three plot we would make if not melted.
