## Series from Lists

```python
# string
country = ['India', 'USA', 'UK', 'Germany', 'France']

# To convert the list to pandas series
print(pd.Series(country)) # here the string is referred as object in pandas.
print()

# integer
batsman = [10, 4, 2, 6, 1]
print(pd.Series(batsman)) # here the integer is referred as int64 in pandas.
```

```python
#  To give custom index to the series, using the index parameter
print(pd.Series(country, index=['a', 'b', 'c', 'd', 'e'])) # here the string is referred as object in pandas.
print()
print(pd.Series(batsman, index=['a', 'b', 'c', 'd', 'e'])) # here the integer is referred as int64 in pandas.
print()
```

```python
# To give a name to the series, using the name parameter
print(pd.Series(country, index=['a', 'b', 'c', 'd', 'e'], name='Countries')) # here the string is referred as object in pandas.
print()
print(pd.Series(batsman, index=['a', 'b', 'c', 'd', 'e'], name='Runs')) # here the integer is referred as int64 in pandas.
```

```python
# Using Dictionary to create a series
country_dict = {'a': 'India', 'b': 'USA', 'c': 'UK', 'd': 'Germany', 'e': 'France'}
print(pd.Series(country_dict, name='Countries')) # here the string is referred as object in pandas.
print()
batsman_dict = {'a': 10, 'b': 4, 'c': 2, 'd': 6}
print(pd.Series(batsman_dict, name='Runs')) # here the integer is referred as int64 in pandas.

# The Key in the dictionary becomes the index of the series and the value becomes the data of the series.
```

## Series Attribute

```python
# size attribute
print(pd.Series(country).size) # size of the series
print(pd.Series(batsman_dict).size)
```

```python
# dtype attribute
print(pd.Series(country).dtype) # dtype of the series
print(pd.Series(batsman_dict).dtype) # dtype of the series
```

```python
# name attribute
print(pd.Series(country).name) # name of the series
print(pd.Series(batsman_dict, name='Runs').name) # name of the series
```

```python
# isunique attribute
print(pd.Series(country).is_unique) # check if the values in the series are unique or
print(pd.Series(batsman_dict).is_unique) # check if the values in the series are unique or not
```

```python
# index attribute in pandas
print(pd.Series(country).index) # index of the series
print(pd.Series(batsman_dict).index) # index of the series
```

```python
# values attribute in pandas
print(pd.Series(country).values) # values of the series
print(pd.Series(batsman_dict).values) # values of the series
print(type(pd.Series(country).values)) # type of the values of the series
```

## Editing Series

```python
# Editing using Index
kohli[1] = 100 # to edit the value at index 1 in the series.
print(kohli)
```

```python
# Editing the index that does not exist will create a new index and add the value to it.
kohli[216] = 200 # to edit the value at index 216 in the series.
print(kohli.tail())
```

```python
# editing using the slicing
kohli[0:5] = 50 # to edit the values from index 0 to 4 in the series.
print(kohli)
```

```python
# editing using fancy indexing
kohli[[1, 3, 5]] = 100 # to edit the values at index 1, 3 and 5 in the series using fancy indexing.
print(kohli)
```

```python
# editing using the boolean masking
kohli[kohli > 50] = 100 # to edit the values in the series that are greater than 50 using boolean masking.
print(kohli)
```

```python
# we can also edit using the label indexing using the loc function but since we have integer index in the kohli series we cannot use the loc function to edit the values in the series.
bollywood["3 Idiots"] = "Shah Rukh Khan" # to edit the value at index "3 Idiots" in the series.
bollywood
```
