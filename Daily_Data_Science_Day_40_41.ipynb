# Feature Scaling

Data Preprocessing contains various processes -
1. Anomally Detection (Outliers, Structure, Dtype, Extra Data)
2. Missing Values
3. Feature Scaling (Scaling the data to fit visualization / Machine Learning Algorithm)
4. Feature Encoding


- `Feature Scaling ->` Feature scaling is the method to limit the range of variables so that they can be compared on common ground.

- `Importance of Feature Scaling ->` It helps algorithms converge faster and models perform better, especially for models like linear regression. logistic regression, and support vector machines.

- `Features ->` The columns that are used and observed to derive the value of another column. [Independant] They are denoted as follows ->

  - X => Features
  - y => Labels

`Methods of Feature Scaling ->`
- Min - Max Scaling
- Standard Scaling (Z Score Normalization)
- Robust Scaling
- Logarithmic Scaling

Most Common library used to perform these scaling is Scikit-Learn / sklearn.

Min-max scaling, also known as rescaling or normalization, is a preprocessing technique used in machine learning to transform the range of features of input data set.
In min-max scaling, each feature of the input data is transformed so that it falls within a specified range, typically between 0 and 1.

After min-max scaling, the range of values of each feature will be between the target range (typically 0 and 1). This method does not reduce the effect of outliers, but it linearly scales them down into a fixed range. The largest occurring data point corresponds to the maximum value and the smallest one corresponds to the minimum value2.
It’s important to note that min-max scaling confines the values to a fixed range unlike standardization.

Standard scaling, also known as standardization, is a preprocessing technique used in machine learning to standardize the range of features of input data set.
In standard scaling, each feature of the input data is transformed so that it has a mean of 0 and a standard deviation of 1. The formula for standard scaling is:
z=x−u​ / s
where:

- x is the feature value,
- u is the mean of the feature values, and
- s is the standard deviation of the feature values1.

So, after standard scaling, the range of values of each feature will depend on how spread out the values were to begin with, but they will be centered around 0 with a standard deviation of 1. This method is sensitive to outliers, and the features may scale differently from each other in the presence of outliers1.
It’s important to note that standard scaling does not confine the values to a fixed range like normalization (which typically scales values to a range of 0-1). The transformed values can be less than 0 and greater than 1.

## Logarithmic Scaling ->

Formula ->

We take the Log10(x)

Logarithmic scaling, or log scaling, is a method used to display numerical data that spans a broad range of values, especially when there are significant differences between the magnitudes of the numbers involved.

Unlike a linear scale where each unit of distance corresponds to the same increment, on a logarithmic scale each unit of length is a multiple of some base value raised to a power, and corresponds to the multiplication of the previous value in the scale by the base value. A logarithmic scale is nonlinear, and as such numbers with equal distance between them such as 1, 2, 3, 4, 5 are not equally spaced. Equally spaced values on a logarithmic scale have exponents that increment uniformly. Examples of equally spaced values are 10, 100, 1000, 10000, and 100000 (i.e., 10^1, 10^2, 10^3, 10^4, 10^5) and 2, 4, 8, 16, and 32 (i.e., 2^1, 2^2, 2^3, 2^4, 2^5).

Exponential growth curves are often depicted on a logarithmic scale to prevent them from expanding too rapidly and becoming too large to fit within a small graph. Some of our senses operate in a logarithmic fashion (Weber–Fechner law), which makes logarithmic scales for these input quantities especially appropriate.

### Min Max Scaling ->

```python
# Code for demonstrating min max scaling

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Sample data
data = {'Value':[10,20,30,40,50]}
df = pd.DataFrame(data)

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(df)

# Print the scaled data
print(scaled_data)
```

```python
# Code for demonstrating Standard Scaling

import pandas as pd
from sklearn.preprocessing import StandardScaler

# Sample data
data = {'Value':[10,20,30,40,50]}
df = pd.DataFrame(data)

# Create a Standard Scaler obj
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(df)

# Print the scaled data
print(scaled_data)
```

```python
# Code to demonstrate Robust Scaling

import pandas as pd
from sklearn.preprocessing import RobustScaler

# Sample data
data = {'Value':[10,20,30,40,50]}
df = pd.DataFrame(data)

# Make a Robust Scaler obj
scaler = RobustScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(df)

# Print the scaled data
print(scaled_data)
```

# Feature Encoding

Feature encoding is the process of converting raw data or text into a format that can be easily understood by machine learning algorithms. Since most algorithms expect numerical input, categorical or text data often needs to be transformed or encoded into numerical format.

There are several techniques for feature encoding:

1. **Mapping**: Binary features can be easily mapped to 0 and 1.

2. **Label Encoding**: Label encoding algorithm is quite simple and it considers an order for encoding, Hence can be used for encoding ordinal data.

3. **One-Hot Encoding**: To overcome the disadvantage of Label Encoding as it considers some hierarchy in the columns which can be misleading to nominal features present in the data. we can use the One-Hot Encoding strategy. One-hot encoding is processed in 2 steps: Splitting of categories into different columns. Put ‘0 for others and ‘1’ as an indicator for the appropriate column.

4. **Frequency Encoding**: We can also encode considering the frequency distribution. This method can be effective at times for nominal features.

5. **Ordinal Encoding**: We can use Ordinal Encoding provided in Scikit learn class to encode Ordinal features.

> Note: One-hot encoding approach eliminates the order but it causes the number of columns to expand vastly. So for columns with more unique values try using other techniques.

## One-Hot Encoding

```python
# Code to demonstrate one-hot encoding

import pandas as pd

# Sample data
data = {'Color':['Red','Green','Blue','Red','Green','Blue']}
df = pd.DataFrame(data)

# Perform one-hot encoding
encoded_data = pd.get_dummies(df, columns=['Color'])

# Print the encoded data
print(encoded_data)
```

## Label Encoding

```python
# Code to demonstrate Label encoding

from sklearn.preprocessing import LabelEncoder

# Sample Data
data = {'Animal':['Dog','Cat','Bird','Dog']}
df = pd.DataFrame(data)

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the data
encoded_data = label_encoder.fit_transform(df['Animal'])

# Print the encoded data
print(encoded_data)
```

## Ordinal Encoding

```python
# Code to demonstrate Ordinal Encoding

from sklearn.preprocessing import OrdinalEncoder

# Sample Data
data = {'Size':['Small','Medium','Large','Small','Medium','Large']}
df = pd.DataFrame(data)

# Create an OrdinalEncoder object
ordinal_encoder = OrdinalEncoder()

# Fit and transform the data
encoded_data = ordinal_encoder.fit_transform(df)

# Print the encoded data
print(encoded_data)
```
