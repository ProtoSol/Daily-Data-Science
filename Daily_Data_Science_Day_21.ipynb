##Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the Dataset

```python
df = sns.load_dataset('titanic')
df
```

```python
df.info()
```

## Dealing with null values

```python
df.isnull().sum()
```

Count Plot or Bar plot

```python
df.survived.value_counts().plot(kind = "bar")
```

```python
sns.countplot(x ='survived', hue = 'sex' ,data = df)
```

## Scatter plot

```python
plt.scatter(data = df, x = 'age', y = 'fare')
```

```python
sns.scatterplot(data = df, x = 'age', y = 'fare', hue = 'pclass')
```

```python
sns.barplot(data = df, x = 'sex', y = 'fare', hue = 'pclass')
```

## Pair Plot by Seaborn

This plots every combination of column that is possible. It takes quite some time to render acc to the length of the feature set.

```python
sns.pairplot(df)
```

## Hist Plot // Histogram

```python
df.hist()
```

```python
df.hist(bins = 10)
```

```python
df.hist(bins = 10, figsize = (9,7))
```

```python
df.hist(bins = 10, figsize = (9,7), grid = False)
```

```python
# For specific Column we do
df["parch"].hist(grid = False)
```

## Box Plot

```python
sns.boxplot(data = df, x = "sex", y = 'fare', hue = "embark_town")
```

## Boxen Plot

```python
sns.boxenplot(data = df, x = "sex", y = 'fare', hue = "embark_town")
```

```python
sns.boxenplot(data = df, x = "sex", y = 'fare', hue = "embark_town")
plt.title("Fare v/s Sex")
```

## Facet Grid by Seaborn

```python
df.columns
```

```python
g = sns.FacetGrid(data = df, hue = "survived", col= "pclass", margin_titles = True,
                 palette = {1:"seagreen", 0:"violet"})
g = g.map(plt.scatter, "fare", "age", edgecolor="black").add_legend()
```

```python
g = sns.FacetGrid(data = df, hue = "survived", col= "sex", margin_titles = True,
                 palette = {1:"seagreen", 0:"violet"})
g = g.map(plt.scatter, "fare", "age", edgecolor="black").add_legend()
```

## Strip Plot

This is basically using scatter instead of bar's area in bar plot.

```python
sns.stripplot(data = df, x = "sex", y = 'fare', hue = "embark_town")
```

## Heat Map

```python
# We will find the correlation of the dataset
corr = df.corr()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corr)
plt.title("Heat Map of Titanic Dataset")
```

## Violin Plot

```python
sns.violinplot(data = df, x = 'sex', y = 'fare', hue = "class")
```
