## Types of Plots ->

1. Scatter Plot

  It is used to plot relationship between two variables. Both values are numerical not categorical.

2. Bar Plot / Chart

  It is used to plot single numerical Variable using the mean and shows the standard deviation scale.

3. Box Plot

  It is used to observe the distribution of data, and possible outliers.

4. Pie Chart / Donut Chart

  As the name suggests it shows the percentage of categorical values.

5. Area Chart

  It shows the data in the form of Area coverage.

6. Spider Plot

  Like a cobweb, it starts from center and spreads out according to the value.

7. Geographical Plot

  It uses Map to plot value on the certain area,

8. Guide to Choose Types of Plot -> [Andrew Abela Chart Guide](https://extremepresentation.typepad.com/files/choosing-a-good-chart-09.pdf)

Polar chart / Radar Chart

## More About Sunburst plot

A `Sunburst Plot` is a type of data visualization that displays hierarchical data as a set of nested circles, representing different levels in the hierarchy. It starts from the center, with the root node, and spreads out according to the value of the corresponding variable. This plot is particularly useful for displaying the composition of a whole, and the relationship of its parts, in a clear and concise way.

```python
# Import the Libraries Required
import plotly.express as px
import seaborn as sns
import pandas as pd

# Load the dataset
df = sns.load_dataset('titanic')

# Create sunburst plot
fig = px.sunburst(
    df,
    path=['sex', 'class', 'who', 'alive', 'alone'],
    values='survived',
    color='sex',
)

# Show the Plot
fig.show()
```

```python
# Sunburst Plot on tips dataset

tips = sns.load_dataset('tips')

fig = px.sunburst(
    tips,
    path=['sex', 'smoker', 'day', 'time'],
    values='total_bill',
    color='sex',
)

fig.show()
```

> It is used to show the heirarchy in the dataset that contains multiple categorical values.
